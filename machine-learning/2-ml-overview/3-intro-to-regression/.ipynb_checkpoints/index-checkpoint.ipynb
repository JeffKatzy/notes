# Introduction to Simple Linear Regression

### Introduction

It's time to explore our first machine learning model, linear regression.  The key word in linear regression is linear - which means line.  In linear regression, we essentially draw a line through our datapoints.  Here we'll see the components of each line.  

Let's explore linear regression by taking another look at how we draw a line through our data.  Let's start with data regarding different daily temperature and the related customers.

| temperature        | actual customers           
| ------------- |:-------------:| 
|    80       | 120 | 
|    65        | 100 | 
|    50        | 85 | 
|    70        | 100 | 
|    60        | 90 | 

````python
temperatures = [80, 65, 50, 70, 60]
amounts = [120, 100, 85, 100, 90]
````

From there, we create a plot of this data.

````python
from graph import plot, trace_values
layout = {'yaxis': {'title': 'customers'}, 'xaxis': {'title': 'temperatures'}}
actual_trace = trace_values(temperatures, amounts, name = 'actual customers')
plot([actual_trace], layout = layout)
````

Then to this, we add our line, which we'll call our linear regression model.

````python
temperatures = [50, 55, 60, 65, 70, 75, 80]
predictions = list(map(lambda temperature: 1.3*temperature, temperatures))
predictions_trace = trace_values(temperatures, predictions, 'lines', name = 'predictions')
plot([actual_trace, predictions_trace], layout = layout)
````

### Dependent and Independent Variables

Now this line represents the formula $customers = 1.3*temperature$.  In our code that plots the chart above, the key place to see this is with the following line: 

```python
predictions = list(map(lambda temperature: 1.3*temperature, temperatures))
```

Now here our formula is: 

$$ customers = 1.3*temperature $$

But this is just an example of a general formula:

$$y = mx$$

We think of $x$ as our input and $y$ as our output.  So as we saw in our example, we can input into our model a different temperature and predict a different number of customers as our output.  

We have previously described $y$ as our **target variable**.  In the context of linear regression, we may also here it described as a **dependent variable** as the output changes depending on a different input -- here it will be 1.3 times the temperature.  

We have previously described $x$ as our **feature variable**.   Another term you may come across is **independent variable**, as it does not depend on anything.  We can plug in any value for $x$ to get an output of $y$.

### Our coefficients

Now let's take a look at that `m` in $y = mx + b$.  This is that number 1.3 in our customers formula.  So $m$ is called a coefficient.  In math, a coefficient just means a number that is multiplied by a variable -- here our independent variable x.  So whatever the value of $x$, we multiply it by our coefficient $m$.

The coefficient determines the slope of the line.  The larger the coefficient, the steeper our line.  Let's see this below where we plot the two linear models:

$$ customers = 1.3*temperature $$

and

$$customers=2∗temperature$$


````python
temperatures = [50, 55, 60, 65, 70, 75, 80]
predictions = list(map(lambda temperature: 1.3*temperature, temperatures))

predictions_trace = trace_values(temperatures, predictions, 'lines', name = 'first model: slope 1.3')

second_model_predictions = list(map(lambda temperature: 3*temperature, temperatures))
second_model_trace = trace_values(temperatures, second_model_predictions, 'lines', name = 'second model: slope 3')
plot([actual_trace, second_model_trace, predictions_trace], layout = layout)
````

If we think about it, it makes sense that the coefficient determines the steepness of our line.

In our second formula, $ customers=3∗temperature$, represented by the green line we see that if the temperature is 50 we predict customers of 150, and with a temperature of 51 we predict 153 customers.  So we are saying that a for every one unit increase in the value of x, the output goes up by 3.  As opposed to in the other model $ customers=1.3∗temperature$, for a one unit increase in the value of x we only increase by 1.3.  So the larger our coefficient, the larger in y our model predicts, per a one unit increase in x.  And accordingly the steeper the slope.

### Y intercept

There's one more component to our line.  This is what'a called our y intercept.  A y intercept is simply the y value when our independent variable, x is zero.  

In our first model above $customers=1.3∗temperature$, if we let our independent variable $temperature = 0$ we calculate $customers = 1.3*0 = 0$.  But this may not be the best model.  What if we want to say that when it is zero degrees outside that it we still predict 30 customers.  We can change our model to be the following: $ customers=1.3∗temperature + 30 $.  So now when the temperature is zero our updated model predicts $ customers=1.3∗0 + 30 = 30$.  

Notice that this also affects every other prediction.  Every previous prediction is increased by 30.  So when it's 50 degrees, we go from predicting 65 customers to 95 customers.  And when it's 51 degrees, we go from predicting 68 customers to 98 customers.  

Notice that by adding in our intercept, we changed the general form of our model from $y = mx $ to $y = mx + b$, where b is our intercept.

Let's plot the models $customers=1.3∗temperature + 30 $ and $customers=1.3∗temperature + 0 $ side by side.

````python
temperatures = [50, 55, 60, 65, 70, 75, 80]
predictions = list(map(lambda temperature: 1.3*temperature, temperatures))

trace_without_intercept_value = trace_values(temperatures, predictions, 'lines', name = 'first model: 1.3x + 0')

### Change this code
second_model_predictions = list(map(lambda temperature: 1.3*temperature + 30, temperatures))


trace_with_intercept = trace_values(temperatures, second_model_predictions, 'lines', name = 'second model: 1.3x + 30')
plot([actual_trace, trace_without_intercept_value, trace_with_intercept], layout = layout)
````

So adding an intercept value, b, raises the line vertically.  While changing our slope m, changes the steepness of the line.  

Try changing the slopes and intercepts in the code above, to see how the line changes, or feel free to add in another line with different slopes and intercepts.

### Summary

In this lesson, we learned about a simple linear regression model.  A simple linear model has one input and one output.  The input is called the independent variable, $x$.  In our example above, customers is the dependent variable and temperature is the independent variable.  This makes sense as the number of customers predicted by our model *depends* on the temperature.  The temperature does not depend on anything in our model, so it's independent.

We also discussed our coefficient, which is the number we multiply our independent variable by.  It is represented by $m$ in the fromula $y = mx $.  We can interpret our coefficient as the impact that our independent variable $x$ has on our dependent variable $y$.  So in the model $ customers = 1.3*temperature$, this means our model predicts that a one degree increase in temperature increases the number of customers by 1.3.  The value of $m$ also determines the slope of our line -- the further $m$ is from zero, the steeper the slope.

Finally we saw another component of our line, the intercept.  The intercept is the predicted output when our independent variable is zero.  So in our model of $ customers = 1.3*temperature + 30 $ we say that even when the temperature is zero, we expect 30 customers.  We also noticed that the intercept increases our predicted output by that value 30, for every input value.  So where we previously to predicted an output of 65 with a temperature of 50, the updated formula predicts 95.  And where we previously to predicted an output of 68 with a temperature of 51, the updated formula predicts 98.
